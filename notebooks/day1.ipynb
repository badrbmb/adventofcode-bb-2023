```python
def replace_spelled_out_digits(in_string, sorted_len=True, reverse_string=False):
    string = copy(in_string)
    words_to_digits = [
        ('one', 1),
        ('two', 2),
        ('three', 3),
        ('four', 4),
        ('five', 5),
        ('six', 6),
        ('seven', 7),
        ('eight', 8),
        ('nine', 9)
    ]
    
    if reverse_string:
        words_to_digits = [(u[::-1], v) for u, v in words_to_digits]
        string = string[::-1]

    if sorted_len:
        words_to_digits = sorted(words_to_digits, key=lambda x: len(x[0]))
    
    # try each digit (ordered by lenght of digit)
    out = ""
    for letter in string:
        out += letter
        for u, v in words_to_digits:
            out = out.replace(u, str(v))
    return out


def get_calibration_value(document: str, un_spell: bool = False) -> int:
    values = document.strip().split('\n')
    pattern = r'\d'
    sum_calibration = 0
    for raw_value in values:
        raw_value = raw_value.strip()
        nums = []
        for reversed_string in [False, True]:
            value = replace_spelled_out_digits(raw_value, reverse_string=reversed_string)
            numbers = re.findall(r"\d", value)
            nums.append(int(numbers[0]))
        calibration_value = int(str(nums[0]) + str(nums[1]))
        sum_calibration +=calibration_value
        print(f"{raw_value=} --> {calibration_value=}")
        
    return sum_calibration
```
